```python
import os
import re
import matplotlib.pyplot as plt
import time
from datetime import datetime
import string
import random
import pandas as pd
import json
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.firefox.options import Options
```

```python
with open('../WTO_data_article/all_article_content.json') as f:
    all_article_content = json.load(f)
```

```python
def check_date_format(raw_date):
    try:
        datetime.strptime(raw_date, "%d %B %Y")
        return True
    except:
        return False
```

```python
months = ['JANUARY', 'FEBRUARY', 'MARCH', 'APRIL', 'MAY', 'JUNE', 'JULY', 'AUGUST', 'SEPTEMBER', 'OCTOBER', 'NOVEMBER', 'DECEMBER']
correct_date_url = set()

for url in all_article_content.keys():
    raw_date = all_article_content[url]['raw_date']

    for m in months:
        if raw_date != None and m in raw_date.upper() and check_date_format(raw_date)==True:
            correct_date_url.add(url)
            all_article_content[url]['date'] = raw_date
            break

incorrect_date_url = set(all_article_content.keys()).difference(correct_date_url)

print('articles with correct date:', len(correct_date_url))
print('articles with incorrect date:', len(incorrect_date_url))
```

```python
def remove_punctuation(input_string):
    return re.sub(r'[^\w\s]', ' ', input_string)
```

```python
has_corrected_cnt = 0
failed_correction = set()

for url in all_article_content.keys():
    # total += 1
    if all_article_content[url]['raw_date'] != None:
        # seperate days lilke dd-dd mm yyyy
        raw_date = remove_punctuation(all_article_content[url]['raw_date'])
        raw_date = raw_date.upper().replace('—', ' ').replace('TH', '').replace('\n', ' ').replace('  ', ' ')
        
    else:
        continue
    
    has_corrected = False
    if check_date_format(raw_date) == False:
        try:
            for m in months:
                if m in raw_date:
                    raw_date_sp = raw_date.split(' ')
                    m_index = raw_date_sp.index(m)
                    if raw_date_sp[m_index-1].isnumeric() and raw_date_sp[m_index+1].isnumeric() and len(raw_date_sp[m_index+1]) == 4:
                        new_date = ' '.join(raw_date_sp[m_index-1:m_index+2])
                        all_article_content[url]['date'] = new_date
                    has_corrected_cnt += 1
                    has_corrected = True
                    break
            if has_corrected == False:
                failed_correction.add(url)
                print(url)
                print(all_article_content[url]['raw_date'], end='\n\n')

        except:
            failed_correction.add(url)
            print(url)
            print(all_article_content[url]['raw_date'], end='\n\n')
        

print(f'has corrected {has_corrected_cnt} articles')

with open('../WTO_data_article/all_article_content.json', 'w') as fp:
    json.dump(all_article_content, fp)
```

```python
# date_extraction_methods
def date_extraction_1(driver):
    introTextDiv = driver.find_element(By.CLASS_NAME, 'introTextDiv')
    date = introTextDiv.find_element(By.CLASS_NAME, 'paralargecolourtext').text
    return date

def date_extraction_2(driver):
    introTextDiv = driver.find_element(By.CLASS_NAME, 'introTextDiv')
    date = ' '.join([kt.text for kt in introTextDiv.find_elements(By.CLASS_NAME, 'kickertext')])
    return date

def date_extraction_3(driver):
    introTextDiv = driver.find_element(By.CLASS_NAME, 'introTextDiv')
    date = ' '.join([pt.text for pt in introTextDiv.find_elements(By.CLASS_NAME, 'paranormaltext')])
    return date

def date_extraction_4(driver):
    introTextDiv = driver.find_element(By.CLASS_NAME, 'introTextDiv')
    date = introTextDiv.find_element(By.TAG_NAME, 'p').text
    return date

def date_extraction_5(driver):
    introTextDiv = driver.find_element(By.CLASS_NAME, 'introTextDiv')
    date = ' '.join([br.text for br in introTextDiv.find_elements(By.TAG_NAME, 'br')])
    return date

def date_extraction_6(driver):
    introTextDiv = driver.find_element(By.CLASS_NAME, 'introTextDiv')
    date = ' '.join(introTextDiv.text.split('\n')[:4])
    return date

def date_extraction_7(driver):
    centerCol = driver.find_element(By.CLASS_NAME, 'centerCol')
    date = centerCol.find_element(By.CLASS_NAME, 'parasmalltext').text
    return date

def date_extraction_8(driver):
    centerCol = driver.find_element(By.CLASS_NAME, 'centerCol')
    date = centerCol.find_element(By.TAG_NAME, 'b').text
    return date

def date_extraction_9(driver):
    centerCol = driver.find_element(By.CLASS_NAME, 'centerCol')
    date = centerCol.find_element(By.TAG_NAME, 'p').text
    return date

def date_extraction_10(driver):
    rightCol = driver.find_element(By.CLASS_NAME, 'rightCol')
    date = datetime.strptime(rightCol.text.split('\n')[3], "%d.%m.%Y").strftime("%d %B %Y")
    return date

date_extraction_methods = [date_extraction_1, date_extraction_2, date_extraction_3, date_extraction_4, 
                           date_extraction_5, date_extraction_6, date_extraction_7, date_extraction_8, 
                           date_extraction_9, date_extraction_10]
```

```python
import os
import matplotlib.pyplot as plt
import time
from datetime import datetime
import string
import random
import pandas as pd
import json
```

```python
# set path WTO_data_article
path = os.path.abspath('../WTO_data_article/')
print(path)
```

```python
for folder in os.listdir(path):
    print(folder)
    record = []

    for file in os.listdir(path + '/' + folder):
        # print(file)
        data = pd.read_csv(path + '/' + folder + '/' + file)
        record.append([''.join(file[:-4].split('_')[2:]), len(data), len(set(data['url']))])
    df = pd.DataFrame(record, columns=['file_name', 'total', 'unique'])
    df.sort_values(by=['file_name'], inplace=True, ascending=True)

    # histogram
    # x = file_name, y = total
    plt.figure(figsize=(20, 5))
    plt.bar(df['file_name'], df['total'])
    plt.xticks(rotation=90)
    plt.xlabel('submenu')
    plt.ylabel('url count')
    plt.title(folder + ' - total url count')
    
```

```python
for folder in os.listdir(path):
    print(folder)
    types = set()

    for file in os.listdir(path + '/' + folder):
        # print(file)
        data = pd.read_csv(path + '/' + folder + '/' + file)
        types = types.union(set(data['type']))
    print(len(types))
    print(types)
```
